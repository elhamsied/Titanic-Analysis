```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.drop("Cabin", axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby("Pclass")["Age"].mean()
```

```python
df["AgeGroup"] = pd.cut(df["Age"],
                       bins=[0, 12, 18, 35, 60, 100],
                       labels=["Child", "Teen", "Young Adult", "Adult", "Senior"])

df.groupby("AgeGroup")["Survived"].mean()
```

```python
df[(df["Sex"] == "female") & (df["Survived"] == 1)]
```

```python
df[(df["Age"] < 12) & (df["Survived"] == 1)]
```

```python
df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
```

INSIGHTS

1. Females were more likely to survive than males.

2. Passenger class affected survival:
   - 1st class had the highest survival rate
   - 3rd class had the lowest survival rate

3. Children were prioritized and had higher survival rates.

4. The highest survival group was:
   Female passengers in 1st class, especially young individuals.
